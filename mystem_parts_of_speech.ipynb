```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
texts = []
texts_names = []
for i in range(7, 13):
    texts_names.append(f"detcorpus ({i}).txt")
    with open(f"/Users/macbook/Downloads/detcorpus ({i}).txt") as f:
        texts.append(f.read())

from nltk import tokenize
sentenced_texts = [tokenize.sent_tokenize(text.replace("\n\n", " ")) for text in texts]
```

```python
from nltk.tokenize import word_tokenize
from pymystem3 import Mystem
m = Mystem()

mystem_texts = [[m.analyze(sent) for sent in text] for text in sentenced_texts]
```

```python
mystem_words = []
for text in range(0,len(mystem_texts)):
    for sent in mystem_texts[text]:
        sent_ind = mystem_texts[text].index(sent)
        for word in sent:
            if len(word) == 2:
                word_ind = sent.index(word)
                if len(word['analysis']) > 0:
                    mystem_words.append((text, sent_ind, word_ind, word))
            else:
                continue
```

```python
mystem_words = pd.DataFrame((mystem_words), columns = ["id_text", "id_sent", "id_word", "word"])
```

```python
import numpy as np
mystem_words['lex'] = [row['word']['analysis'][0]['lex'] 
                       if "lex" in row['word']['analysis'][0].keys() else np.nan
                       for index, row in mystem_words.iterrows()]
mystem_words['wt'] = [row['word']['analysis'][0]['wt'] 
                      if "wt" in row['word']['analysis'][0].keys() else np.nan
                      for index, row in mystem_words.iterrows()]
mystem_words['gr'] = [row['word']['analysis'][0]['gr'] 
                      if "gr" in row['word']['analysis'][0].keys() else np.nan
                      for index, row in mystem_words.iterrows()]
```
